Explain and list some differences between

_Answer:_ They are all used for storage.

RAM is used as temporary storage for data that a computer needs to access quickly. It stores the information that is actively being used or processed by the CPU. When the computer is turned off, the data in RAM is lost, so it has volatile memory. On the other hand, Solid State Drives (SSD) provide long-term storage of data and are used to store the operating system, applications, and files.

Unlike RAM, SSDs retain data when the power is turned off (non-volatile storage). They are used as a replacement for HDDs in many modern computers, but they are more expensive than HDDs. They are built using flash memory (NAND flash) and provide the same block-oriented interface as disk storage.

Hard Drives (HDD), also called magnetic disk drives, are also used for long-term data storage and are non-volatile. They are mechanical devices, and data access requires a read–write head to move to the required cylinder, and the rotation of the platters must then bring the required sector under the read–write head. Magnetic disks thus have a high latency for data access compared to SSDs.

## W2

With regards to disk drives, define

- Seek time
- Rotational latency time
- Transfer time/data transfer rate


_Answer:_ Accessing data on a specific sector of a disk requires the disk's arm to first align over the correct track before waiting for the disk's rotation to bring the sector beneath it.

- Seek time: This is the duration needed to move the arm into position over the right track, a period that lengthens with the arm's travel distance. Typically, seek times fall between 2 to 20 milliseconds, influenced by the track's distance from the arm's starting point. Smaller disks often feature shorter seek times as the distance the head must cover is less.

- Rotational latency time: After the head aligns with the intended track, the delay before the targeted sector rotates to a position under the head is known as rotational latency time. Modern disks rotate at speeds from 5400 RPM (90 rotations per second) to 15,000 RPM (250 rotations per second), translating to a rotational latency ranging from 4 milliseconds to 11.1 milliseconds per rotation. Disks with faster rotational speeds are preferred for minimizing latency in certain applications.

- Transfer time/data transfer rate: The process of transferring data starts when the disk's head positions over the initial sector of the requested data. The data transfer rate measures how fast data is written to or read from the disk, with contemporary disk systems capable of peak transfer speeds between 50 to 200 megabytes per second. However, the actual transfer rates for the disk's inner tracks can be much lower than the maximum due to having fewer sectors per track. For instance, a disk advertised with a peak transfer rate of 100 megabytes per second might only sustain about 30 megabytes per second on its innermost tracks.

## W4

Define and list some benefits of

- Fixed-length records
- Variable-length records
- Row-oriented storage
- Column-oriented storage

The predictability of record size can lead to more consistent and predictable performance in data processing and retrieval operations.

_Answer:_

- Fixed-length vs Variable-length records:

In fixed-length records, each record is of the same size, meaning that the amount of space allocated to each field within a record is predetermined and constant across all records. This structure is simple and straightforward, facilitating uniformity. On the other hand, variable-length records do not have a predefined size. Instead, the space allocated for each record can expand or contract based on the actual size of the data it contains. These records are often managed through the use of delimiters or length indicators.

Data retrieval and manipulation, such as insertion and deletion, are simpler to implement for fixed-length records because the space made available by a deleted record is exactly the space needed to insert a record, unlike variable-length records. However, variable-length records offer efficient use of storage space as the space is allocated based on the actual size of the data, which can significantly reduce wasted space compared to fixed-length records, especially in databases with wide variations in data size. The performance of fixed-length records in data processing and retrieval operations is more consistent and predictable, but variable-length records are more flexible, accommodating a wide range of data sizes, making them suitable for storing record types that contain repeating fields such as arrays or multisets, and the presence of multiple record types within a file

- Row-oriented vs Column-oriented storage:

In traditional databases, data is stored in a row-oriented format, where all attributes of a tuple are stored together, and tuples are stored sequentially in a file. Conversely, column-oriented storage stores each attribute of a relation in a separate file, with values from successive tuples stored at successive positions in the file. This approach can offer two advantages:
1) It can be efficient in data retrieval: Column-oriented storage allows for only the needed attributes to be fetched for queries involving a few fields from a table with many attributes, avoiding the unnecessary loading of irrelevant data. This is in stark contrast to row-oriented storage, where adjacent, unneeded attributes might also be loaded, wasting memory and processing resources.
2) It can improve data compression: By grouping similar data types together, column-oriented storage enhances data compression, reducing storage costs and speeding up data retrieval.

However, the main downside of column-oriented storage compared to row-oriented storage is the increased I/O operations required to store or fetch a complete record, making it less ideal for transactional processing. Despite this, its benefits make it a preferred choice for data warehousing tasks focused on scanning and aggregating vast amounts of data rather than handling individual records.

## W5

Explain and list some differences between

- RAID 0
- RAID 1
- RAID 5


_Answer:_

Redundant Arrays of Independent Disks (RAID) are disk-organization techniques that aim to achieve improved performance, reliability, and storage redundancy. By organizing data across multiple disks, RAID configurations can offer faster data access, increased data reliability, or a combination of both.

- **RAID 0**, refers to disk arrays with striping at the level of blocks, where data is divided into blocks and each block is written to a separate disk drive. This method enhances performance because it allows multiple disks to read and write data simultaneously. However, RAID 0 lacks redundancy—meaning it does not provide any fault tolerance. This configuration is best suited for situations where speed is crucial and data loss is not a concern.

- **RAID 1**, refers to disk mirroring with block striping, which involves copying identical data onto two or more drives. This setup provides redundancy; if one drive fails, the data can be retrieved from another, unlike in RAID 0. While RAID 1 offers excellent data protection, its cost-effectiveness is lower because it requires double the storage capacity to store the same amount of data. RAID 1 is suitable for applications requiring high reliability.

- **RAID 5** uses block-interleaved distributed parity to provide a balance between performance, storage efficiency, and data protection. In RAID 5 configurations, both data and parity information (used for data recovery in case of a disk failure) are distributed across all disks in the array. For each set of data blocks, a corresponding parity block is stored on a different disk. This setup ensures that if any single disk fails, the lost data can be reconstructed from the remaining disks using the parity information. RAID 5 is ideal for environments that require a good balance between high performance, efficient storage utilization, and reliable data protection.

## Person ID Function

- `personID` is formed using the following rules:

1. The ID consists of three sections: `[lastSubstr][firstSubstr][number]`
2. `lastSubstr` is formed by lowercasing `nameLast`, then taking the first 5 letters. If `nameLast` is less than 5 letters, use the entire `nameLast`.
3. `firstSubstr` is formed by lowercasing `nameFirst`, then taking the first 2 letters. If `nameFirst` is less than 2 letters, use the entire `nameFirst`.
4. For a specific combination of `[lastSubstr][firstSubstr]`, `number` starts from 1 and increments. `number` should be padded to have length 2.
5. `nameFirst` and `nameLast` may contain periods ".", hyphens "-", and spaces " ". You should remove these characters from `nameFirst` and `nameLast` **before** doing the above substring processing.

- As an example, starting from an empty table, below is what `personID` would be assigned to the following names (assuming they were inserted in the order that they are shown)

| nameFirst | nameLast | personID  |
|-----------|----------|-----------|
| Donald    | Ferguson | fergudo01 |
| David     | Aardsma  | aardsda01 |
| Doe       | Fergue   | fergudo02 |
| J. J.     | Park     | parkjj01  |

- Write a SQL function that generates a person ID using the above rules
    - You should determine what parameters and return type are needed
    - This function will be called by triggers in the next section. **It is up to you which logic you put in the function and which logic you put in the triggers.**
        - That is, if you plan to place the bulk of your logic in your triggers, then your function could be a few lines.
    - You may define helper functions
    - You may add additional attributes to your table if it helps

```python
%%sql

DROP FUNCTION IF EXISTS GetNextNumber;

create function GetNextNumber(lastSubstr VARCHAR(5), firstSubstr VARCHAR(2)) returns INT
    reads sql data
begin
    declare maxNumber INT DEFAULT 0;

    SELECT MAX(CAST(
        SUBSTRING(personID, LENGTH(lastSubstr) + LENGTH(firstSubstr) + 1) AS UNSIGNED))
        INTO maxNumber
    FROM People
    WHERE personID LIKE CONCAT(lastSubstr, firstSubstr, '%');

    IF maxNumber IS NULL THEN
        SET maxNumber = 0;
    END IF;

    RETURN maxNumber + 1;
end;
```

```python
%%sql

DROP FUNCTION IF EXISTS compute_next_personID;

create
    definer = root@localhost function s24_hw3.compute_next_personID(nameFirst varchar(100),
                                                                    nameLast varchar(100))
                                                                    returns varchar(9)
    reads sql data
begin
    declare personID VARCHAR(9);
    declare lastSubstr VARCHAR(5);
    declare firstSubstr VARCHAR(5);
    declare number INT DEFAULT 1;
    
    SET nameFirst = REPLACE(REPLACE(REPLACE(REPLACE(nameFirst, '.', ''), '-', ''),
                                                                ' ', ''), "'", '');
    SET nameLast = REPLACE(REPLACE(REPLACE(REPLACE(nameLast, '.', ''), '-', ''),
                                                                ' ', ''), "'", '');

    SET lastSubstr = LOWER(LEFT(nameLast, 5));

    SET firstSubstr = LOWER(LEFT(nameFirst, 2));

    SET number = GetNextNumber(lastSubstr, firstSubstr);

    SET personID = CONCAT(lastSubstr, firstSubstr, LPAD(number, 2, '0'));

    RETURN personID;
end;

```

```python
%%sql

DROP TRIGGER IF EXISTS set_people;
DROP TRIGGER IF EXISTS update_people;

create definer = root@localhost trigger s24_hw3.set_people
    before insert
    on s24_hw3.People
    for each row
begin
    set NEW.personID = s24_hw3.compute_next_personID(NEW.nameFirst, NEW.nameLast);
end;

create definer = root@localhost trigger s24_hw3.update_people
    before update
    on s24_hw3.People
    for each row
begin
  if old.personID != new.personID THEN
    signal sqlstate '50000'
    set MESSAGE_TEXT = "You cannot change the personID. This is a warning. Ignored.",
                                                                  mysql_errno = 1001;
    set new.personID = old.personID;
  elseif (OLD.nameFirst != NEW.nameFirst) OR (OLD.nameLast != NEW.nameLast) then
    set NEW.personID = s24_hw3.compute_next_personID(NEW.nameFirst, NEW.nameLast);
  END IF;
end;

```

## Create and Update Procedures

- You must implement two stored procedures

1. `createPerson(nameFirst, nameLast, birthYear, birthCountry, deathYear, deathCountry, personID)`
    1. `personID` is an out parameter. It should be set to the ID generated for the person.
    2. All the other parameters are in paramaters
2. `updatePerson(personID, nameFirst, nameLast, birthYear, birthCountry, deathYear, deathCountry, newPersonID)`
    1. `newPersonID` is an out parameter. It should be set to the ID of the person after the update (even if it didn't change).
    2. All the other parameters are in parameters.
        1. `personID` is used to identify the row that the user wants to update. The other in parameters are the values that the user wants to set.
        2. **Ignore null in parameters.** Only update an attribute if the in parameter is non-null.
        
- Depending on how you implemented your triggers, these procedures could be as simple as calling `insert`/`update` and setting the out parameters

```python
%%sql

DROP PROCEDURE IF EXISTS createPerson;
DROP PROCEDURE IF EXISTS updatePerson;

CREATE PROCEDURE createPerson(
    IN nameFirst VARCHAR(100),
    IN nameLast VARCHAR(100),
    IN birthYear INT,
    IN birthCountry VARCHAR(100),
    IN deathYear INT,
    IN deathCountry VARCHAR(100),
    OUT newPersonID VARCHAR(9)
)
BEGIN
    INSERT INTO People (nameFirst, nameLast, birthYear, birthCountry, deathYear, deathCountry)
    VALUES (nameFirst, nameLast, birthYear, birthCountry, deathYear, deathCountry);

    SET @nameFirst = REPLACE(REPLACE(REPLACE(REPLACE(nameFirst, '.', ''), '-', ''),
                                                                ' ', ''), "'", '');
    SET @nameLast = REPLACE(REPLACE(REPLACE(REPLACE(nameLast, '.', ''), '-', ''),
                                                                ' ', ''), "'", '');

    SET @lastSubstr = LOWER(LEFT(@nameLast, 5));
    SET @firstSubstr = LOWER(LEFT(@nameFirst, 2));

    SELECT personID INTO newPersonID FROM People
    WHERE personID LIKE CONCAT(@lastSubstr, @firstSubstr, '%')
    ORDER BY personID DESC
    LIMIT 1;
END;


CREATE PROCEDURE updatePerson(
    IN oldPersonID VARCHAR(9),
    IN newNameFirst VARCHAR(100),
    IN newNameLast VARCHAR(100),
    IN newBirthYear INT,
    IN newBirthCountry VARCHAR(100),
    IN newDeathYear INT,
    IN newDeathCountry VARCHAR(100),
    OUT newPersonID VARCHAR(9)
)
BEGIN
    SELECT nameFirst INTO @nameFirst FROM People WHERE personID = oldPersonID;
    SELECT nameLast INTO @nameLast FROM People WHERE personID = oldPersonID;

    UPDATE People
    SET
        nameFirst = IF(newNameFirst IS NOT NULL, newNameFirst, nameFirst),
        nameLast = IF(newNameLast IS NOT NULL, newNameLast, nameLast),
        birthYear = IF(newBirthYear IS NOT NULL, newBirthYear, birthYear),
        birthCountry = IF(newBirthCountry IS NOT NULL, newBirthCountry, birthCountry),
        deathYear = IF(newDeathYear IS NOT NULL, newDeathYear, deathYear),
        deathCountry = IF(newDeathCountry IS NOT NULL, newDeathCountry, deathCountry)
    WHERE personID = oldPersonID;

    SET @nameFirst = REPLACE(REPLACE(REPLACE(REPLACE(IF(newNameFirst IS NOT NULL, newNameFirst,
                                            @nameFirst), '.', ''), '-', ''), ' ', ''), "'", '');
    SET @nameLast = REPLACE(REPLACE(REPLACE(REPLACE(IF(newNameLast IS NOT NULL, newNameLast,
                                            @nameLast), '.', ''), '-', ''), ' ', ''), "'", '');

    SET @lastSubstr = LOWER(LEFT(@nameLast, 5));
    SET @firstSubstr = LOWER(LEFT(@nameFirst, 2));

    SELECT personID INTO newPersonID FROM People
    WHERE personID LIKE CONCAT(@lastSubstr, @firstSubstr, '%')
    ORDER BY personID DESC
    LIMIT 1;

END;
```

## DV0

- This question is an example of what is required from you
- Create a bar graph showing the amount of time each season ran for (in seconds)
- You should use the `episodes_scenes` table


- Note: `season_running_time <<` in the following cell saves the output of the SQL query into a local Python variable `season_running_time`

```python
%%sql

season_running_time <<

with one as (
    select seasonNum, episodeNum, sceneNum, sceneEnd, time_to_sec(sceneEnd) as sceneEndSeconds,
           sceneStart,  time_to_sec(sceneStart) as sceneStartSeconds,
            time_to_sec(sceneEnd)-time_to_sec(sceneStart) as sceneLengthSeconds
    from episodes_scenes
),
two as (
    select seasonNum, episodeNum, max(sceneEnd) as episodeEnd, max(sceneEndSeconds)
                                                                as episodeEndSeconds
    from one
    group by seasonNum, episodeNum
),
    three as (
        select seasonNum, cast(sum(episodeEndSeconds) as unsigned) as totalSeasonSeconds,
               sec_to_time(sum(episodeEndSeconds)) as totalRunningTime
        from two
        group by seasonNum
    )
select * from three;
```

```python
# You must show the visualization

season_running_time[['seasonNum', 'totalSeasonSeconds']].plot.bar(x='seasonNum',
                                                                  y='totalSeasonSeconds')
```

## DV2

- Create a bar chart showing the number of episodes that every location (not sublocation) appeared in
    - You are counting the number of episodes, not scenes. If a location appeared in multiple scenes in a single episode, that should increment your count only by one.
    - You should order your chart on the number of episodes descending, and you should only show the top 10 locations
- You should use the `episodes_scenes` table
- As an example, your bar chart may look like this:

<img src="./images/NP3.png">

```python
%%sql

location_episode_count <<

SELECT 
    sceneLocation, 
    COUNT(*) AS episodeCount
FROM (
    SELECT DISTINCT
        sceneLocation,
        seasonNum,
        episodeNum
    FROM episodes_scenes
) AS distinct_episodes
GROUP BY sceneLocation
ORDER BY episodeCount DESC
limit 10;
```

## DV3

- Create a scatter plot showing the relationship between the number of episodes (not scenes) a character appears in and their screen time (in seconds)
    - A character's screen time is the sum of the time lengths of all the scenes that the character appears in
- You should use the `episodes_characters` and `episodes_scenes` tables
- As an example, your scatter plot may look like this:

<img src="./images/NP4.png">

```python
%%sql

episode_count_screen_time <<

select 
    ec.charactername, 
    count(distinct ec.episodenum, ec.seasonnum) as episodecount,
    sum(timestampdiff(second, str_to_date(es.scenestart, '%H:%i:%s'),
                    str_to_date(es.sceneend, '%H:%i:%s'))) as screentimeseconds
from 
    episodes_characters ec
join 
    episodes_scenes es on ec.episodenum = es.episodenum and ec.seasonnum = es.seasonnum
                                                        and ec.scenenum = es.scenenum
group by 
    ec.charactername;

```

## DV4

- Create a bar chart showing the number of exclusive characters in each season
    - An exclusive character is a character that appeared in only that season, no other season
    - You should order your chart on the number of exclusive characters descending
- You should use the `episodes_characters` table
    - You can assume `characterName` is unique across all characters. That is, a single name is one unique character.
- As an example, your bar chart may look like this:

<img src="./images/NP5.png">

```python
%%sql

season_exclusive_characters <<

select 
    seasonnum,
    count(charactername) as exclusivecharactercount
from
    (select 
        charactername, 
        min(seasonnum) as seasonnum
    from 
        episodes_characters
    group by 
        charactername
    having 
        count(distinct seasonnum) = 1) as unique_characters
group by 
    seasonnum
order by 
    exclusivecharactercount desc;
```
